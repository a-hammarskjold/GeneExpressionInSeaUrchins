```
import scanpy as sc
```

```
sc.settings.verbosity = 3
sc.logging.print_header()
sc.settings.set_figure_params(dpi=80, facecolor='white')
```

```
lv2 = sc.read('/content/drive/My Drive/DSC 541/541 - Gene Expression/lv2hpf_anno.csv',cache=True).transpose()
lv3 = sc.read('/content/drive/My Drive/DSC 541/541 - Gene Expression/lv3hpf_anno.csv',cache=True).transpose()
lv4 = sc.read('/content/drive/My Drive/DSC 541/541 - Gene Expression/lv4hpf_anno.csv',cache=True).transpose()
lv5 = sc.read('/content/drive/My Drive/DSC 541/541 - Gene Expression/lv5hpf_anno.csv',cache=True).transpose()
lv6 = sc.read('/content/drive/My Drive/DSC 541/541 - Gene Expression/lv6hpf_anno.csv',cache=True).transpose()
lv7 = sc.read('/content/drive/My Drive/DSC 541/541 - Gene Expression/lv7hpf_anno.csv',cache=True).transpose()
lv8 = sc.read('/content/drive/My Drive/DSC 541/541 - Gene Expression/lv8hpf_anno.csv',cache=True).transpose()
lv9 = sc.read('/content/drive/My Drive/DSC 541/541 - Gene Expression/lv9hpf_anno.csv',cache=True).transpose()
lv10 = sc.read('/content/drive/My Drive/DSC 541/541 - Gene Expression/lv10hpf_anno.csv',cache=True).transpose()
lv11 = sc.read('/content/drive/My Drive/DSC 541/541 - Gene Expression/lv11hpf_anno.csv',cache=True).transpose()
lv12 = sc.read('/content/drive/My Drive/DSC 541/541 - Gene Expression/lv12hpf_anno.csv',cache=True).transpose()
lv13 = sc.read('/content/drive/My Drive/DSC 541/541 - Gene Expression/lv13hpf_anno.csv',cache=True).transpose()
lv14 = sc.read('/content/drive/My Drive/DSC 541/541 - Gene Expression/lv14hpf_anno.csv',cache=True).transpose()
lv15 = sc.read('/content/drive/My Drive/DSC 541/541 - Gene Expression/lv15hpf_anno.csv',cache=True).transpose()
lv16 = sc.read('/content/drive/My Drive/DSC 541/541 - Gene Expression/lv16hpf_anno.csv',cache=True).transpose()
lv18 = sc.read('/content/drive/My Drive/DSC 541/541 - Gene Expression/lv18hpf_anno.csv',cache=True).transpose()
lv20 = sc.read('/content/drive/My Drive/DSC 541/541 - Gene Expression/lv20hpf_anno.csv',cache=True).transpose()
lv24 = sc.read('/content/drive/My Drive/DSC 541/541 - Gene Expression/lv24hpf_anno.csv',cache=True).transpose()
lv24_2 = sc.read('/content/drive/My Drive/DSC 541/541 - Gene Expression/lv24hpf_2_anno.csv',cache=True).transpose()
```

```
sc.pp.filter_cells(lv2, min_genes=200)
sc.pp.filter_genes(lv2, min_cells=10)
 
#Printing a box-whisker plot of the cells
sc.pl.highest_expr_genes(lv2)
 
#Identifying and pulling mitochondrial cells
lv2.var['mt'] = lv2.var_names.str.startswith('MT-')
 
#Calculating metrics
sc.pp.calculate_qc_metrics(lv2, qc_vars=['mt'], percent_top=None, log1p=False, inplace=True)
 
#Printing a violin plot of the metrics
sc.pl.violin(lv2, ['n_genes_by_counts', 'total_counts', 'pct_counts_mt'], jitter=0.4, multi_panel=True)
 
#Separating out highly variable genes
lv2=lv2[lv2.obs.n_genes_by_counts < 2500, : ]
lv2=lv2[lv2.obs.pct_counts_mt < 5, : ]
sc.pp.normalize_total(lv2, target_sum=1e4)
sc.pp.log1p(lv2)
sc.pp.highly_variable_genes(lv2, min_mean=0.0125, max_mean=3, min_disp=0.5)
sc.pl.highly_variable_genes(lv2)
 
#regressing and identifying variance ratios
lv2 = lv2[: , lv2.var.highly_variable]
sc.pp.regress_out(lv2, ['total_counts', 'pct_counts_mt'])
sc.pp.scale(lv2, max_value=10)
 
#Computing Principal Component Analysis (PCA)
sc.tl.pca(lv2, svd_solver='arpack')
sc.pl.pca_variance_ratio(lv2, log=True)
 
#Identifying Neighbors
sc.pp.neighbors(lv2, n_neighbors=10, n_pcs=40) #can change this
 
#Computing UMAP
sc.tl.umap(lv2,n_components=3)
sc.tl.louvain(lv2)
sc.tl.leiden(lv2)
sc.pl.umap(lv2, color=['louvain','leiden'])
```
